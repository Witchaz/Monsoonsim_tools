```python
import pandas as pd
import numpy as np
```

# ทำการ Assmble ข้อมูลทั้งหมดเป็น df เดียวกัน

```python
retail_sale_df = pd.read_csv('RETAIL-sale.csv')
retail_inventory_df = pd.read_csv('RETAIL-inventory.csv')
marketing_marketing_df = pd.read_csv('MARKETING-marketing.csv')
forecast_forecast_df = pd.read_csv('FORECAST-retail_demand.csv')
```

## เอา Space ออกจาก Column name แล้ว drop Column ที่ไม่ใช้ออก 

```python
retail_inventory_df.columns = retail_inventory_df.columns.str.strip()
retail_inventory_df.set_index("Day",inplace=True)
retail_inventory_df
```

```python
retail_sale_df.columns = retail_sale_df.columns.str.strip()
retail_sale_df.drop(columns=["COGS", "Revenue"], inplace=True)
retail_sale_df.set_index("Day",inplace=True)
retail_sale_df
```

```python
marketing_marketing_df.columns = marketing_marketing_df.columns.str.strip()
marketing_marketing_df.drop(["Type"],axis=1,inplace=True)
marketing_marketing_df.set_index("Day",inplace=True)
marketing_marketing_df
```

```python
forecast_forecast_df.columns = forecast_forecast_df.columns.str.strip()
# forecast_forecast_df.drop(["Category"], inplace=True,axis=1)
forecast_forecast_df.rename(columns={"Category":"Day"},inplace=True)
forecast_forecast_df.set_index("Day",inplace=True)
forecast_forecast_df
```

## ทำการ Merge

```python
df = forecast_forecast_df.merge(retail_sale_df,on="Day").merge(marketing_marketing_df,on="Day").merge(retail_inventory_df,on="Day")
df.drop(["Dept_x","Dept_y","Product_x","Product_y"],inplace=True,axis=1)
# df.sort_index(inplace=True)
df
```
