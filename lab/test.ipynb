```python
import matplotlib as mpl
import pandas as pd
import numpy as np

import random
```

```python

df = pd.read_csv("./csv/FORECAST-retail_demand.csv")
df.columns = df.columns.str.strip()
df
```

```python
# df = pd.read_csv(b2c_file)
# ลบคอลัมน์ 'Category' ออกเพราะไม่ได้ใช้ในการคำนวณ
df = df.drop(columns=["Category"])

# แยกชื่อคอลัมน์เพื่อสร้าง MultiIndex โดยแบ่งเป็น 'City' และ 'Product'
new_columns = pd.MultiIndex.from_tuples(
    [col.split('-', 1) for col in df.columns], names=["City", "Product"]
)
df.columns = new_columns
for i in new_columns.get_level_values(0).drop_duplicates(keep='first'):
    print(i)
# คำนวณ mean, max, min ตาม City และ Product
result = df.agg(['mean', 'max', 'min'])

result
```
