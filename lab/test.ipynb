```python
import matplotlib as mpl
import pandas as pd
import numpy as np

import random
```

```python

df = pd.read_csv("MARKETING-Survey.csv")
df.columns = df.columns.str.strip()
df
```

```python

def condition(v):
    if v > 10:
        return "Hot"
    elif v > 5:
        return "normal"
    else:
        return "ignore"

def make_pretty(styler):
    styler.set_caption("Marketing survey")
    styler.background_gradient(axis=None, vmin=5, vmax=15, cmap="RdPu")
    return styler

```

```python

df['Age Group'] = pd.cut(df['Age'], bins=[0, 30, 60, float('inf')], labels=['30 and below', '30-60', 'Above 60'])
media_counts_in_groups = df.groupby(['Age Group','Product Preference'])['Media Preference'].value_counts().unstack(fill_value=0)
media_counts_in_groups.style.pipe(make_pretty)
media_counts_in_groups.reset_index(inplace=True)
media_counts_in_groups
```
